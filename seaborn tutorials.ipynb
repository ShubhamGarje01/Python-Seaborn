```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flight = pd.read_csv("flights.csv")
flight.head()
```

```python
sns.relplot(x = 'passengers', y = 'month', data = flight)
```

```python
sns.relplot(x = 'passengers', y = 'month', hue = 'year', data = flight)
```

```python
tips = sns.load_dataset("tips")
sns.relplot(x = 'time', y = 'tip', kind = 'line', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'boxen', data = tips)
```
